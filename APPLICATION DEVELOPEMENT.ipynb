```python


from tkinter import *
import mysql.connector
from mysql.connector import Error


conn = mysql.connector.connect(host='localhost',database='register', user='root',password='admin',port=3307)

global cursor

if conn.is_connected():
    print('connected to mysql database')

root = Tk()
root.geometry('500x500')
root.title("Registration Form")


highway_mpg=IntVar()
price=IntVar()
var = IntVar()
c=StringVar()
var1= IntVar()



def database():
   name1=highway_mpg.get()
   email=price.get()
   gender=var.get()
   country=c.get()
   prog=var1.get()

   cursor = conn.cursor()
   cursor.execute('INSERT INTO form VALUES(%s,%s,%s,%s,%s)',(name1,email,gender,country,prog))
   conn.commit()
   print('insert successfully')
   cursor.close()
   conn.close()
    
   
   
             
label_0 = Label(root, text="Automobile Details Form",width=20,font=("bold", 20))
label_0.place(x=90,y=53)


label_1 = Label(root, text="highway_mpg",width=20,font=("bold", 10))
label_1.place(x=80,y=130)

entry_1 = Entry(root,textvar=highway_mpg)
entry_1.place(x=240,y=130)

label_2 = Label(root, text="price",width=20,font=("bold", 10))
label_2.place(x=68,y=180)

entry_2 = Entry(root,textvar=price)
entry_2.place(x=240,y=180)

label_3 = Label(root, text="fuel-type",width=20,font=("bold", 10))
label_3.place(x=70,y=230)

Radiobutton(root, text="gas",padx = 5, variable=var, value=1).place(x=235,y=230)
Radiobutton(root, text="diesel",padx = 20, variable=var, value=2).place(x=290,y=230)

label_4 = Label(root, text="make",width=20,font=("bold", 10))
label_4.place(x=70,y=280)

list1 = ['alfa-romero','audi','bmw','chevrolet','dodge','isuzu','honda','jaguar','mercedes-benz','nisson','mercury','volvo'];

droplist=OptionMenu(root,c, *list1)
droplist.config(width=15)
c.set('select') 
droplist.place(x=240,y=280)

label_4 = Label(root, text="num-of-doors",width=20,font=("bold", 10))
label_4.place(x=85,y=330)
var2= IntVar()
Checkbutton(root, text="two", variable=var1).place(x=235,y=330)

Checkbutton(root, text="four", variable=var2).place(x=290,y=330)

Button(root, text='Submit',width=20,bg='brown',fg='white',command=database).place(x=180,y=380)

root.mainloop()

```
